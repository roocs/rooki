# Demo for cmip6-decadal intake catalog


```python
import intake
```

## Open remote catalog

```python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
```

## Load catalog for c3s-cmip6-decadal

```python
df_cmip6_decadal = cat['c3s-cmip6-decadal'].read()
df_cmip6_decadal.head()
```

```python
institutions_ids = list(set(df_cmip6_decadal.institution_id))
institutions_ids
```

## Check member_id per institution

```python
for institution_id in institutions_ids:
    print()
    print(institution_id)
    df_inst = df_cmip6_decadal.loc[
        (df_cmip6_decadal.institution_id==institution_id)
    ]
    member_years = list(set(df_inst.member_year))
    member_years.sort()

    for member_year in member_years:
        df = df_inst.loc[
            (df_inst.member_year==member_year)
        ]
        ensembles = list(set(df.member_ensemble))
        ensembles.sort()
        ensembles = [ensemble for ensemble in ensembles if "i1" in ensemble]
        print(member_year, ensembles)

```
