They noticed (a larger version of) the following request failing:
```
dataset = "projections-cmip6"
request = {
    "temporal_resolution": "monthly",
    "experiment": "historical",
    "variable": "daily_minimum_near_surface_air_temperature",
    "model": "ukesm1_0_ll",
    "month": ["01"],
    "year": ["1980"]
}
```


## Search Catalog

````python
import intake
````

````python
cat_url = "https://raw.githubusercontent.com/cp4cds/c3s_34g_manifests/master/intake/catalogs/c3s.yaml"

cat = intake.open_catalog(cat_url)
list(cat)
````

````python
df_cmip6 = cat['c3s-cmip6'].read()
df_cmip6.head()
````

````python
df = df_cmip6.loc[
    (df_cmip6.source_id=="UKESM1-0-LL")
    & (df_cmip6.table_id=="Amon")
    & (df_cmip6.variable_id=="tasmin")
    & (df_cmip6.experiment_id=="historical")
]
df
````

````python
ds_ids = list(df.ds_id.unique())
ds_ids
````

## Run subset operator

````python
import os
os.environ['ROOK_URL'] = 'http://rook.dkrz.de/wps'

from rooki import rooki
````

````python
resp = rooki.subset(
    collection=ds_ids,
    time='1980/1980',
    time_components='year:1980|month:jan',
)
resp.ok
````

````python
resp
````
